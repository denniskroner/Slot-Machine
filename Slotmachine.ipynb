## Class Purse
This class defines the gaming account of an instance.
It is only used to take money from the account and 
put it back in. It is also used to return the amount available

```
class purse
    initialise the class taking the instance as only arguments
        set the default account balance equal to 10
        use an instance variable called balance

    define the method to credit the profit taking instance and profit as arguments
        add profit to account balance

    define the method to debit the stake taking instance and stake as arguments
        subtract stake from account balance

    define the method to display the account balance taking only instance as argument
        return account balance
```

````python
class Purse():
    #initialises the class object and sets the default value 
    #for the attribute equal to 10€
    #each instance starts the game with 10€
    def __init__(self):
        self.__balance = 10
    
    #defines the method credit to credit the profit to the account  
    def credit(self, profit):
        self.__balance += profit
    
    #defines the method debit to debit money from the account to place a bet
    def debit(self, stake): 
        self.__balance -= stake
    
    #defines the method get_balance to return the balance of the account
    def get_balance(self):
        return self.__balance
    
````

## Class Slot
This is a sub-class or the parent class 'Column' and includes the actual game.
It allows an instance to bet an amount of money, pull the handle of the slot machine in order to launch the game.
It also includes methods to show the symbols or faces of each round and return the score of a round.

```
class Slot
    initialise the class taking instance and the slot symbols as arguments
        inherit the initialisation method from the parent class column

    define the method that takes the bet of an instance with taking instance and stake as arguments
        store the stake in the instance variable stake

    define the method that pulls the handle of the slot machine and takes only the instance as argument
        call the function from the class 'Column' that changes the 3 symsbols of the game randomly
        store the result in an intance variable like face_res

    define the method that shows the faces for each round taking the instance as only argument
        print the 3 faces stored in the instance variable face_res

    define the method that calculates the score of a round taking only instance as argument
        store the 3 faces in a different variable e.g. res 
        if the 3 element in the array res are equal then
            print out the score (2 times the stake)
            and return the score
        else if only two faces are equal then
            print out the result (1.5 times the stake)
            and return the result
        else
            print out that the player scored 0 and return 0
```

````python
class Slot():
    def __init__(self,face):
        Column.__init__(self,face)
    
    #defines method take_bet which stores the wager of a player(instance)
    def take_bet(self, stake):
        self.__stake = stake
    
    #defines the method pull_handle to initiate the round
    def pull_handle(self):
        #call method change_faces to play the round
        self.__face_res = Column.change_faces(self)
    
    #defines the method show_slot to print the faces of the slot
    def show_slot(self):
        print(*self.__face_res)
    
    #defines the method score_slot to evalute the amount of money the player won
    def score_slot(self):
        res = self.__face_res
        
        #if all three faces are equal player wins 100% of his bet
        if(res[0] == res[1] == res[2]):
            print("\nYou Score %.1f - " %(self.__stake * 2), end = " ")
            return self.__stake * 2
        #if two faces are equal player wins 50% of his bet
        elif ((res[0] == res[1]) or (res[0] == res[2]) or (res[1] == res[2])):
            print("\nYou Score %.1f - " %(self.__stake * 1.5), end = " ")
            return self.__stake * 1.5
        #three different faces -> player lost the stake
        else:
            print("\nYou Score 0 - ", end = " ")
            return 0
````

## Class Column

````python
class Column():
    #initialise the class with attribute face
    def __init__(self, face):
        #store the 3 diffenrent faces for the game
        self.__face = face
    
    #defines the method change_face for playing a round    
    def change_faces(self):
        #returns 3 faces chosen randomly
        return (choice(self.__face),choice(self.__face),choice(self.__face))
  
````

## Function validate_bet
This function validates in the bet made by a player is valid. That means that each bet 
is bigger than or equal to 2 and smaller or equal to the account balance. It also checks if the player enter 'N' to leave the game and if other characters were entered which are not permitted.
It returns the valid user input. That could be the stake or 'N' for leaving the game.
```
define a function that check the validity of the user input taking the account balance as argument
    initiate a variable by setting it equal to True in order to enter the while loop
    
    while loop as long as user input is invalid
          prompt user for input and store it in variable stake
          
          remove white spaces from variable
          
          if stake is equal to 'N' then 
              set invalid_bet equal to False and leave the loop
          else if check if user input is numerical if yes then
              if user input is greater or equal to the minimum bet and smaller or equal to the balance then
                  set invalid_bet equal to False and leave the loop      
    return the stake
```

````python
def validate_bet(balance):
    """This function takes account balance as input and returns the stake."""
    
    #initialise to enter loop
    invalid_bet = True
    #while loop until the user inout is valid
    while invalid_bet:
        #prompt user for input
        stake = input("How much do you bet: ")
        
        #remove whitespaces
        stake = stake.strip()
        
        #if evalutaes the user input
        if stake == "N": invalid_bet = False
        elif stake.isnumeric():
            if int(stake) >= 2 and int(stake) <= balance: invalid_bet = False
        
    return stake
````
